```python
# интерполяция
from skimage.transform import rescale
import numpy as np
import matplotlib.pyplot as plt
astronaut = data.astronaut()
rescaled = rescale(astronaut,[2,2,1],order = 0)
astronaut = astronaut[:20,:20,:]
hue_rotations = np.linspace(0, 1, 6)
fig, axes = plt.subplots(nrows=2, ncols=3,figsize=(10, 10), sharex=True, sharey=True)
array = [0,1,2,3,4,5]
i = 0
for ax, hue in zip(axes.flat, hue_rotations):
    rescaled = rescale(astronaut,[2,2,1],order = array[i])
    ax.imshow(rescaled)
    i+=1
fig.tight_layout()
print(axes.flat, hue_rotations)
```
